```python
import requests

# Gene name
gene_name = input("Enter a gene name: ")

# Construct URL for HGNC API endpoint for the gene name
url = f"https://rest.genenames.org/fetch/symbol/{gene_name}"

# Set headers for HTTP GET request
headers = {
    "Accept": "application/json"
}

# Send HTTP GET request to HGNC API
response = requests.get(url, headers=headers)

# Parse JSON response from HGNC API
data = response.json()

# Extract aliases for the gene
aliases = data["response"]["docs"][0]["alias_name"]

# Print aliases for the gene
print("Aliases for " + gene_name + ": " + ", ".join(aliases))
```
